## Marketing and Retail Analytics

In today’s competitive e-commerce landscape, efficient inventory management is crucial for minimizing costs and maximizing profitability. OList, an emerging e-commerce company, has recently faced losses due to high inventory costs and inefficient stock management. To address this challenge, this project focuses on optimizing inventory by analyzing sales data, identifying top revenue-generating products, and understanding customer purchase behavior through Market Basket Analysis.

The project is structured as follows:

* Data Exploration & Cleaning: Identifying and addressing missing values, duplicates, and data inconsistencies to ensure high-quality analysis
* Data Visualization: Analyzing the most ordered products by quantity and revenue to determine high-demand and slow-moving products.
* Market Basket Analysis: Applying association rule mining to uncover frequently purchased product combinations and improve cross-selling strategies.
* Dashboard & Insights: Developing an interactive dashboard to visualize key trends and summarizing findings with actionable recommendations to optimize inventory and reduce unnecessary costs.

By implementing these data-driven strategies, this project aims to enhance OList’s inventory planning, improve operational efficiency, and drive profitability.

#### Calculate RFM Metrics

```python
# Recency
# most recent order date is used to calculate Recency

referance_date = merge4['order_purchase_timestamp'].max()
referance_date
```

```python
# Created new column >> Sum of price + shipping charges
merge4['total_value'] = merge4['price'] + merge4['shipping_charges']
```

```python
rfm_product = merge4.groupby('product_id').agg (
           
        Recency = ('order_purchase_timestamp', lambda x: (referance_date-x.max()).days),
        Frequency = ('order_id', 'nunique'),
        Monetary=('total_value', 'sum')          # Sum of price + shipping charges
    
                                               ).reset_index().sort_values(by= ['Recency', 'Frequency', 'Monetary'])


rfm_product.head()
```

```python
# Assign RFM Scores
rfm_product['R_score'] = pd.cut(rfm_product['Recency'], bins=4, labels=[4,3,2,1])    # Lower is better
rfm_product['F_score'] = pd.cut(rfm_product['Frequency'], bins=4, labels=[1,2,3,4])  # Higher is better
rfm_product['M_score'] = pd.cut(rfm_product['Monetary'], bins=4, labels=[1,2,3,4])   # Higher is better

rfm_product['Final_Score'] = rfm_product['R_score'].astype(int) + rfm_product['F_score'].astype(int) + rfm_product['M_score'].astype(int)

rfm_product_sorted = rfm_product.sort_values (by= 'Final_Score', ascending=False)
rfm_product_sorted
```
